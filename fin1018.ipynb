```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

def y(p):
  x,y=p
  return (x**2+y**2-16,-x**2+2*x+4-y)
x,y=fsolve(y,(1,0))
print (x,y)

l=np.linspace(-10,10,100000)
p=np.sqrt(16-l**2)
plt.plot(l,p,'b')
plt.plot(l,-p,'b')
e=np.linspace(-10,10,100)
n=-e**2+2*e+4
plt.plot(e,n)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()

```
