```python
import numpy as np
import matplotlib.pyplot as plt


# Определяем функцию и её производную

def f(x):
    return x**3 - 5*x**2 - 4.5

def f_prime(x):
    return 3*x**2-10*x

# Метод Ньютона
def newton_method(initial_guess, tolerance=1e-7, max_iterations=100):
    x_current = initial_guess
    history = [x_current]  # Сохраняем историю для графика

    for _ in range(max_iterations):
        x_next = x_current - f(x_current) / f_prime(x_current)
        history.append(x_next)

        if abs(x_next - x_current) < tolerance:
            break

        x_current = x_next

    return x_next, history

# Параметры метода
initial_guess = -2  # Начальное приближение
root, history = newton_method(initial_guess)

# Вывод результата
print(f"Найденный корень: {root}")

# Визуализация
x_values = np.linspace(-2, 8, 400)
y_values = f(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='f(x) = $x^3 - 5x^2 - 4.5$', color='blue')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(root, color='red', lw=0.5, ls='--', label=f'Корень: {root:.4f}')

# Отметим каждую итерацию метода Ньютона
for i in range(len(history)):
    plt.plot(history[i], f(history[i]), 'ro')  # Точки итераций

plt.title('Метод Ньютона для уравнения $x^3 - 5x^2 - 4.5 = 0$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```
