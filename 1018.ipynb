```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
def result(x):
  return (x**2+5*x*np.sin(x)-10)
x=fsolve(result,-3)
o=np.linspace(-20,20,100)
plt.plot(o,result(o))
plt.scatter(x,0,color='red')
plt.show
print(x)
print(result(x))
```
