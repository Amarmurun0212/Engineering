```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from mpl_toolkits import mplot3d
df = pd.read_excel('/content/Plotsss.xlsx')
fig = plt.figure()
x = df['Plot']
y = df['PlotTej']
z = df['Eff']
fig = plt.figure(figsize =(14, 14))
ax = plt.axes(projection ='3d')
surf = ax.plot_trisurf(x, y, z, linewidth=0,   antialiased=False,cmap='viridis', edgecolor='none')
ax.set_xlabel('Плотность угля')
ax.set_xlim(1000, 1800)
ax.set_ylabel('Плотность тяжёлой среды')
ax.set_ylim(1000, 1400)
ax.set_zlabel('Эффективность')
ax.set_zlim(0,80)
plt.show()
ax.view_init(20,20)
fig
```
